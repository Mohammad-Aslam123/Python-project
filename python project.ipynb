```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
df=pd.read_excel("employees stats.xlsx")
```

```python
df.head()
```

```python
df['Height']=np.random.randint(150,181,size=len(df))
```

```python
df=pd.DataFrame(df)
df.to_excel("employees stats.xlsx")
df
```

```python
#EXERCISE 1

total_employees=df['Name'].count()

team_count=df['Team'].value_counts()

percentage_splitting=(team_count/total_employees)*100

# new_details=p({'Team': team_count.index, 'Number of employees': team_count.values, 'percentage split':percentage_splitting.values})
print(total_employees,team_count.count,percentage_splitting)

#PIE CHART
plt.figure(figsize=(8,10))
plt.pie(team_count.values,labels=team_count.index,autopct='%1.1f%%')
plt.title('Percentage Splitting of Employees in Each Team')
plt.show()
```

```python
#EXERCISE 2
position_count=df['Position'].value_counts()
print('Number of employees each position',position_count)

#BAR CHART
plt.bar(position_count.index,position_count.values,color='green')
plt.title('Number of employees each position')
plt.xlabel('position')
plt.ylabel('number of employees')
plt.show()
```

```python
#EXERCISE 3
age=df['Age'].value_counts()
print('which age group most of the employees belong to',age)

plt.hist(df['Age'],bins=[0,10,20,30,40,50],edgecolor='black')
plt.title('Distribution of Employees Across Age Groups')
plt.xlabel('age')
plt.ylabel('number of employees')
plt.show()
```

```python
#EXERCISE 4

max_salary_spending = df.groupby(['Team', 'Position'])['Salary'].sum().idxmax()
print(max_salary_spending)
```

```python
#EXERCISE 5

#SCATTERPLOT
sns.scatterplot(x='Age',y='Salary',hue='Position',data=df)
plt.title('Scatterplot of age vs salary')

plt.show()

#HEAT MAP
new=df[['Age','Salary']]
corr=new.corr()
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('correlation matrix of age and salary')
plt.show()

corr_coefficient=df['Age'].corr(df['Salary'])
print('coefficient of correlation between age and salary=',corr_coefficient)
```
